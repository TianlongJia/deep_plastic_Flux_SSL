# Zip each folder from one path (to many zip files)

```python
import os
import zipfile

# Define the path where the folders are located
base_path = r'U:\AIMMW\Paper_4\Flux\datasets\Amsterdam\tiles_224\1'

# Loop through each item in the base_path
for folder_name in os.listdir(base_path):
    folder_path = os.path.join(base_path, folder_name)
    
    # Check if the item is a directory
    if os.path.isdir(folder_path):
        # Define the path for the output zip file
        zip_filename = os.path.join(base_path, f"{folder_name}.zip")
        
        # Create a new zip file and add the folder contents to it
        with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
            for root, _, files in os.walk(folder_path):
                for file in files:
                    # Get the full file path
                    full_path = os.path.join(root, file)
                    # Add the file to the zip archive, using a relative path
                    zipf.write(full_path, os.path.relpath(full_path, folder_path))

print("All folders have been zipped successfully.")
```

# Zip some folders in a list

```python
import os
import zipfile

# Define the path where the folders are located
folder_list = [
               r"U:\AIMMW\Paper_4\Flux\datasets\GV\Final_ds\GV_pretrain_20per",
               r"U:\AIMMW\Paper_4\Flux\datasets\GV\Final_ds\GV_pretrain_60per",
               r"U:\AIMMW\Paper_4\Flux\datasets\GV\Final_ds\GV_pretrain_100per"
               ]

for folder in folder_list:
        # Define the output zip file name based on the folder name
        folder_name = os.path.basename(folder.rstrip('/\\'))
        output_zip = os.path.join(os.path.dirname(folder), f"{folder_name}.zip")

        # Create the zip file for the current folder
        with zipfile.ZipFile(output_zip, 'w', zipfile.ZIP_DEFLATED) as zipf:
            # Walk through each folder and add files to the zip archive
            for root, dirs, files in os.walk(folder):
                for file in files:
                    # Create the file's full path
                    file_path = os.path.join(root, file)
                    # Add the file to the zip, preserving its relative folder structure
                    zipf.write(file_path, os.path.relpath(file_path, folder))
        print(f"Zipped {folder} to {output_zip}")
```

# Upzip each zip file from one path (to many folder)

```python
import os
import zipfile

# Define the path where the zip files are located
base_path = r'/scratch/tjian/Data/Flux/images_pretrain_300K/train/'

# Loop through each item in the base_path
for item in os.listdir(base_path):
    # Construct the full path to the item
    item_path = os.path.join(base_path, item)
    
    # Check if the item is a zip file
    if item.endswith('.zip'):
        # Define the directory to extract the contents
        extract_folder = os.path.join(base_path, item[:-4])  # Remove the .zip extension
        
        # Create the directory if it doesn't exist
        if not os.path.exists(extract_folder):
            os.makedirs(extract_folder)
        
        # Unzip the file into the specified directory
        with zipfile.ZipFile(item_path, 'r') as zip_ref:
            zip_ref.extractall(extract_folder)
    print(item, " is unzipped")

print("All zip files have been extracted successfully.")
```
